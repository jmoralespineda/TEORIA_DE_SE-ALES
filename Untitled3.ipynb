```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# reproducibilidad
torch.manual_seed(42)

# parámetros
T = 1000
N = 64
t = torch.arange(T).float()

# 1) Generar ensemble de señales (sinus + ruido)
signals = []
for i in range(N):
    # variar fase/forma entre señales para realismo (reducida para mayor estacionariedad)
    raw = torch.sin(0.02 * t + 0.01 * i) + 0.5 * torch.sin(0.20 * t + 0.003 * i) \
          + 0.25 * torch.randn(T)
    signals.append(raw)
signals = torch.stack(signals)   # shape (N, T)
```

```python
# 2) Normalizar cada señal por su valor absoluto máximo -> rango aproximado [-1,1]
max_per_signal = signals.abs().amax(dim=1, keepdim=True)   # (N,1)
norm_signals = signals / (max_per_signal + 1e-12)          # evita división por 0

# señales de interés (normalizadas, sin offsets)
first = norm_signals[0].clone()   # primera señal normalizada en [-1,1]
last  = norm_signals[-1].clone()  # última señal normalizada en [-1,1]
```

```python
# 3) Escalado/offset para mostrar en la figura:
#    - queremos que 'last' esté entre [-1, 1]
#    - queremos que 'first' esté desplazada hacia arriba entre ~[2, >3]
last_off = last * 0.98           # tope en ≈[-0.98, 0.98] (prácticamente [-1,1])
first_off = first * 0.9 + 2.8    # escala ±0.9 y offset 2.8 -> rango ≈[1.9, 3.7]

# imprimir rangos para verificar
print("Rangos comprobación:")
print(f" first_off  min/max = {first_off.min().item():.3f} / {first_off.max().item():.3f}")
print(f" last_off   min/max = {last_off.min().item():.3f} / {last_off.max().item():.3f}")
```

```python
# 4) Estadísticos (sobre señales normalizadas, sin offsets)
mean_signal = norm_signals.mean(dim=0)
std_signal  = norm_signals.std(dim=0)

all_samples = norm_signals.flatten()
mu_ens = all_samples.mean().item()
sigma_ens = all_samples.std().item()

# time-averaged (ej. de la primera señal normalizada)
mu_time = first.mean().item()
sigma_time = first.std().item()
```

```python

# 5) PDFs (evaluadas sobre coordenada vertical 'y')
ymin, ymax = -1.2, 4.0
y_vals = np.linspace(ymin, ymax, 500)
pdf_ens = (1.0 / (sigma_ens * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((y_vals - mu_ens)/sigma_ens)**2)
pdf_time = (1.0 / (sigma_time * np.sqrt(2*np.pi))) * np.exp(-0.5 * ((y_vals - mu_time)/sigma_time)**2)
```

```python
# 6) Posición y escala de las PDFs (a la derecha)
x_pos = T + 120      # coordenada x donde "pegan" las PDFs
width = 200.0        # ancho horizontal de la PDF

# Dibujado
plt.figure(figsize=(14,6))

# señales (azul arriba y azul claro abajo)
plt.plot(t.numpy(), first_off.numpy(), color='blue', lw=1.2, alpha=1.0,
         label=f'First Signal (1/{N})\n(μ={first.mean().item():.3f}, σ={first.std().item():.3f})')
plt.plot(t.numpy(), last_off.numpy(),  color='blue', lw=1.2, alpha=1.0,
         label=f'Last Signal ({N}/{N})\n(μ={last.mean().item():.3f}, σ={last.std().item():.3f})')

# mean y std (sobre señales normalizadas, sin offsets)
plt.plot(t.numpy(), mean_signal.numpy(), 'red', linestyle='--', lw=1.8, label='Mean Signal')
plt.plot(t.numpy(), std_signal.numpy(),  'green', linestyle='-.', lw=1.6, label='Std Signal')

# líneas horizontales negras como guía (baseline central y baseline superior)
plt.axhline(y=0.0, color='black', linewidth=1.0)
plt.axhline(y=2.8, color='black', linewidth=1.0)

# línea vertical magenta para marcar la región de las PDFs (como en la figura)
plt.axvline(x=T, color='magenta', linewidth=1.5)

# PDFs rotadas a la derecha (ensemble = magenta, time-avg = cyan)
plt.plot(x_pos + pdf_ens * width, y_vals, color='darkmagenta', lw=3.5, label=f'Ensemble-averaged Gaussian PDF\n(μ={mu_ens:.3f}, σ={sigma_ens:.3f})')
plt.plot(x_pos + pdf_time * width + 6, y_vals, color='darkcyan', lw=3.5, label=f'time-averaged Gaussian PDF\n(μ={mu_time:.3f}, σ={sigma_time:.3f})')

# estética
plt.xlabel('Time')
plt.ylabel('Amplitude / Probability Density')
plt.title('First, Last, Mean, Std, and Gaussian PDFs of Normalized Signals')
plt.grid(alpha=0.3)
plt.ylim([ymin, ymax])
plt.xlim([0, x_pos + width + 50])
plt.legend(loc='upper right', fontsize=9)
plt.tight_layout()
plt.show()
```
